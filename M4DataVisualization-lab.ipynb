# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
df = pd.read_sql_query("SELECT * FROM master", conn)
df
df['ConvertedComp'].plot(kind='hist', bins=10)
plt.title("Histogram of Converted Compensation")
```

### Box Plots


Plot a box plot of `Age.`


```python
df['Age'].plot(kind='box', vert=False)
plt.title("Box Plot of Age")
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
df.plot(x='Age', y='WorkWeekHrs', kind='scatter', figsize=(10,6))
plt.title("Scatter Plot of Age and WorkWeekHrs")
plt.xlabel("Age")
plt.ylabel("WorkWeekHrs")
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import plotly.express as px
QUERY="""
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master
"""
bubble_df=pd.read_sql_query(QUERY,conn)

#Replace NaN values with the mean
mean_age=bubble_df['Age'].mean()
bubble_df['Age'].fillna(mean_age, inplace=True) 

mean_crh=bubble_df['CodeRevHrs'].mean()
bubble_df['CodeRevHrs'].fillna(mean_crh, inplace=True)

#Normalize Age
norm_age = (bubble_df['Age'] - bubble_df['Age'].min()) / (bubble_df['Age'].max() - bubble_df['Age'].min())

fig = px.scatter(bubble_df, x='WorkWeekHrs', y='CodeRevHrs', size= norm_age, size_max=60)
fig.update_layout(
    title='Bubble Plot of WorkWeekHrs and CodeRevHrs in relation to Age',
    xaxis_title='WorkWeekHrs',
    yaxis_title='CodeRevHrs'
)
fig.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
table_nameDB = 'DatabaseDesireNextYear'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_nameDB)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])

QUERY = """
SELECT DatabaseDesireNextYear,COUNT(*) as count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by COUNT(*)
"""
df_DB= pd.read_sql_query(QUERY,conn)

df_DB.tail()

top_5_DB= df_DB.nlargest(5, 'count')

#Create a pie chart
fig = px.pie(top_5_DB, values='count', names='DatabaseDesireNextYear', title='Top 5 Databases Respondents Wish to Learn')
fig.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY="""
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master WHERE Age BETWEEN 30 AND 35
group by Age
order by Age
"""
stacked_df=pd.read_sql_query(QUERY,conn)

#Replace NaN values with the median
stacked_med_age=stacked_df['Age'].median()
stacked_df['Age'].fillna(stacked_med_age, inplace=True) 

stacked_med_crh=stacked_df['CodeRevHrs'].median()
stacked_df['CodeRevHrs'].fillna(stacked_med_crh, inplace=True)
stacked_df_age=stacked_df.groupby('Age', axis=0).median()

#Plot stacked bar chart
stacked_df_age.plot(kind='bar', figsize=(10, 6), stacked=True)

plt.xlabel('Age') # add to x-label to the plot
plt.ylabel('Hours') # add y-label to the plot
plt.title('Median Hours by Age') # add title to the plot

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY="""
SELECT ConvertedComp,Age
FROM master WHERE Age BETWEEN 45 AND 60
"""
Line_df=pd.read_sql_query(QUERY,conn)
Line_df

Line_df.set_index('Age',inplace=True)
Line_df.dropna(subset=['ConvertedComp'], inplace=True)

order=['ConvertedComp']
Line_df.groupby('Age')[order].median().plot(kind='line', figsize=(10, 6))

plt.xlabel('Age') # add to x-label to the plot
plt.ylabel('Converted Comp') # add y-label to the plot
plt.title('Median Compensation by Age') # add title to the plot

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY="""
SELECT MainBranch, COUNT(*)
FROM master
group by MainBranch
"""
branch_df=pd.read_sql_query(QUERY,conn)
branch_df

branch_df.plot(kind='barh', x='MainBranch')
```
